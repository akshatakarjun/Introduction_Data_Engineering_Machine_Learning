```python
# import dependencies
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1.	Read the iris dataset using Pandas (ref Option 2 above)
2.	Extract the Iris-setosa and Iris-verginica samples only
3.	Extract the petal length and petal width. These shall be our features.

```python
import pandas as pd
df = pd.read_csv('iris_data.csv', header=0)

# reading the first 5 lines of data set
df.head()
```

```python
# reading the last five lines of dataset
df.tail()
```

```python
# reading the unique values in the class column
df[' class'].unique()
```

```python
# extract rows which have iris setosa and virginica
cond1 = df[' class'] == 'Iris-setosa'
cond2 = df[' class'] == 'Iris-virginica'

# df.loc[row,col]

data_row = df.loc[(cond1) | (cond2),:]
```

```python
# extracting petal length and width column data
data_col = data_row.iloc[:,2:5]
data_col
```

4.	Plot the dataset as a scatter plot with the two Iris types shown as different colours

```python
#  plotting the two categories against their petal length and width using the seaborn library

sns.scatterplot(data = data_col, x = ' petal_length', y = ' petal_width', hue = ' class')
```

5.	Use the perceptron in sklearn.linear_model to train on the Iris data previously used. Set verbose = 1; have a look at the epochs, it converges very quickly?

```python
# So as not to confuse with our previously defined Perceptron class
from sklearn.linear_model import Perceptron as Perceptron
#creating a perceptron model and printing the training progress using vrbose
model = Perceptron(verbose= 1)
model.fit(data_col.iloc[:,:2],data_col.iloc[:,2])
```

6.	Using many random inputs (e.g. 5000) plot to see if you can see the decision boundary for the feature scaled case

```python
import random

# generating random data
np.random.seed(0)

random_petal_width = np.random.uniform(min(data_col[' petal_width']), max(data_col[' petal_width']),5000)
random_petal_length = np.random.uniform(min(data_col[' petal_length']), max(data_col[' petal_length']), 5000)

random_data = pd.DataFrame({' petal_length': random_petal_length,' petal_width': random_petal_width } )
```

```python
random_data.head()
```

```python
# adding the random data as a new column
random_data['predicted_val'] = model.predict(random_data)
```

```python
# scatter plot to see the distribution of data
sns.scatterplot(data = random_data, x = ' petal_length', y = ' petal_width', hue = 'predicted_val')
```

#### Commonality and Differences between artificial neurons and biological neurons
##### > Artificial neurons as model of biological neurons
##### > Are they a good model?

A biological neuron is a nerve cell that sends message to different parts of the body. Once the neuron reaches a threshold value, it sends a message to the brain nucleus to carry out certain actions depending on the importance of the task. The neuron consists of dendrites, nucleus and axons wherein the dendrites receives signal from other cells and is paased to the nucleus which processes the information and in turn, fires an output through axons.

Screenshot 2023-10-09 024705.png

A similar concept can be used in the field of Artificial Neural Network. ANN as a mathematical model is inspired by the biological neuron. A perceptron can be seen as a single unit of an Artificial Neural Network. A simplified model of biological neuron & simulates the behaviour of one cell.

The input data for a perceptron, which can be raw data or outputs from other perceptrons is the same as the input signal of a neuron in the brain. There are two parameters in a neural network model - weights and bias. The input data is multiplied by weights which are numerical values and they control the level of importance of each input data. While the bias helps to shift the activation function of each perceptron to a non-zero value because if one of the weights/inputs is zero, there is no computation. The optimal values of these parameters are obtained during the learning(training) process. The perceptron takes these input values and replicates just like the nucleus except that the perceptron has a linear function and activation function which is non linear. The linear function computes and processes to model linear relationships in the data.

An activation function helps in modeling real-time data which might have non-linear relationships between the attributes.
Firing a neuron - The perceptron fires/activates only when the values of the linear function(z) exceeds the threshold value(as set by the user). The type of activation function (f) determines how the neuron fires and the bias ( dash line inside the perceptron) controls the ease of firing. If the input signals reaches a certain threshold then it gives out an output signal of either 1 or 0.

The perceptron trains the model with each sample training data, learning rate and update weight.





The generic operations of both the perceptron and biological neuron is similar in a way that they take inputs, process it, fires an  output upon raching a threshold.
However, unlike the brain, they do not learn on their own and cannot replica the brain by tring to retain/recall the information.
